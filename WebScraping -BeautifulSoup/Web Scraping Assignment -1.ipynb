```python
#import necessary libraries
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
#url from webscraping
url="https://simplilearn.com/resources"
```

```python
#scape web page content
webpage=urlopen(url)
```

```python
#create soup object
soup=BeautifulSoup(webpage,'html.parser')
webpage.close()
```

```python
#view content
print(soup.contents)
```

```python
#prettify to print in html
print(soup.prettify())
```

```python
#view the title of the scrapped document
print(soup.title)
```

```python
#find all href link present in the webpage
for href in soup.findAll("a",href=True):
    print(href['href'])
```

```python
#view all the header names
for article in soup.findAll('h2'):
    print(article.string)
```

```python
for article_topic in soup.findAll('h4'):
    print(article_topic.string)
```
