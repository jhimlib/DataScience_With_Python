<h2>Task1: Which movies have maximum views/ratings?</h2>

```python
df=df_movies.fillna(value=0)
#transpodf.head()
```

```python
#Transpose and reset Index
df1=df.T
df1=df1.reset_index()
df1


df1.columns = df1.iloc[0]
df1 = df1.reset_index()
df1 = df1.drop(df.index[0])
df1 = df1.reset_index()
df1 =df1.drop(["level_0","index"],axis = 1)
df1.head()

```

```python
#add 2 new columns:averageval,MaxVal at the end
df1=df1.assign(averageval=df1.iloc[:,1:206].mean(axis=1))
df1=df1.assign(MaxVal=df1.iloc[:,1:206].sum(axis=1))
df1.rename(columns={'user_id':'Movie'}, 
                 inplace=True)
df1.head()
```

```python
#Create a new dataframe of moview and average val and max val
df_new=df1[['Movie','averageval','MaxVal']]
df_new.head()
```

```python
#Data Insight:
#Which movies have maximum views/ratings? Movie16
print("Which movies have maximum views/ratings?")
print(df_new.loc[df_new['MaxVal'].idxmax()]['Movie'])
```

```python
#What is the average rating for each movie?
df_new.iloc[:,0:2].head()
```

```python
#Define the top 5 movies with the maximum ratings.
df_new.nlargest(5, 'MaxVal').iloc[:,0:3]
```
