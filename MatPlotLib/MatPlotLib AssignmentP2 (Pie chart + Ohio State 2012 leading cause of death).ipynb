```python
#import necessary libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Data from ohio state, leading causes of death in the state from year 2012
#cause of deaths
cause='Chronic Disease','Unintentional Injuries','Alzheimers','Influenza and Pneumonia','Sepsis','Others'
percentile=[62,5,4,2,1,26]
```

```python
plt.figure(figsize=(10,10))
explode=(0.05,0,0,0,0,0)
plt.pie(percentile,labels=cause,explode=explode,startangle=70)
plt.axis('equal')
plt.title('Ohio State -2012: Leading Causes of Death')
plt.show()
```
